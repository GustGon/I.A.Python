```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("csv/kc_house_data.csv")

# Exibir o cabeçalho do arquivo
df.head()
```

```python
cdf = df[['id','price','bedrooms','bathrooms','sqft_living','zipcode','yr_built','grade','sqft_above','sqft_basement','floors',
          'sqft_lot','waterfront','view']]
cdf.head(9)
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['id','bedrooms','bathrooms','sqft_living','zipcode','yr_built','grade','sqft_above','sqft_basement','floors',
          'sqft_lot','waterfront','view']])
y = np.asanyarray(train[['price']])
regr.fit (x, y)
# Os coeficientes
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['id','bedrooms','bathrooms','sqft_living','zipcode','yr_built','grade','sqft_above','sqft_basement','floors',
          'sqft_lot','waterfront','view']])
x = np.asanyarray(test[['id','bedrooms','bathrooms','sqft_living','zipcode','yr_built','grade','sqft_above','sqft_basement','floors',
          'sqft_lot','waterfront','view']])
y = np.asanyarray(test[['price']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Variância = 1: Predição perfeita
print('Variance score: %.2f' % regr.score(x, y))
```
